```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
FHS_data = pd.read_excel("C:/Users/erueb/dev/python-projects/STAT400-project-one/Framingham_data.xlsx")
```

```python
# Part 1
# Examining gender and coronary heart disease fate to see if a relationship exists

genders = ["Male", "Female"]
# Defining conditions that we will need to look for
male = FHS_data["sex"] == 1
totNumMales = len(FHS_data[male])

female = FHS_data["sex"] == 2
totNumFemales = len(FHS_data[female])

noCHD = FHS_data["chdfate"] == 0
numNoCHD = len(FHS_data[noCHD])

hasCHD = FHS_data["chdfate"] == 1
numHasCHD = len(FHS_data[hasCHD])


print("Number of males: {}".format(totNumMales))
print("Number of females: {}".format(totNumFemales))

# Double check that 0 means NOT disease and 1 means HAS disease
print("Number of people NOT diagnosed with coronary heart disease: {}".format(numNoCHD))
print("Number of people diagnosed with coronary heart disease: {}".format(numHasCHD))
```

```python
# Descriptive Stats for Males
```

```python
# Histogram for males
numMalesNoCHD = len(FHS_data[male & noCHD])
proportionMaleNoCHD = numMalesNoCHD/totNumMales
numMalesWithCHD = len(FHS_data[male & hasCHD])
proportionMaleHasCHD = numMalesWithCHD/totNumMales

proportionCHDArrM = [proportionMaleNoCHD, proportionMaleHasCHD]
maleCHDGroupsArr = ["No CHD", "Diagnosed with CHD"]

y_pos = np.arange(2)
plt.bar(y_pos, proportionCHDArrM, align='center')
plt.xticks(y_pos, maleCHDGroupsArr)
plt.ylabel('Proportion of all males in study')
plt.title('Relative frequency of male CHD')
axes = plt.gca()
axes.set_ylim([0, 1])
plt.show()
```

```python
# Histogram for females
numFemalesNoCHD = len(FHS_data[female & noCHD])
proportionFemaleNoCHD = numFemalesNoCHD/totNumFemales
numFemalesWithCHD = len(FHS_data[female & hasCHD])
proportionFemaleHasCHD = numFemalesWithCHD/totNumFemales

proportionCHDArrF = [proportionFemaleNoCHD, proportionFemaleHasCHD]
femaleCHDGroupsArr = ["No CHD", "Diagnosed with CHD"]

y_pos = np.arange(2)
plt.bar(y_pos, proportionCHDArrF, align='center')
plt.xticks(y_pos, maleCHDGroupsArr)
plt.ylabel('Proportion of all females in study')
plt.title('Relative frequency of female CHD')
axes = plt.gca()
axes.set_ylim([0, 1])
plt.show()
```

```python
# Begin Parts 2 and 3
# Function to generate the histograms for parts 2 and 3
def generateProportionHistogram(title: str, indVar: str, xlabel: str):
    sbp = FHS_data[indVar]
    sbpSortedNoDupes = sbp.sort_values().drop_duplicates().dropna()
    chdPercentages = []
    binnedSortedNoDupes = pd.cut(sbpSortedNoDupes, bins=6, precision = 2)
    bins = pd.Series(binnedSortedNoDupes).drop_duplicates()
    # Calculate proportions
    for interval in bins:
        numHavingCHDForInterval = 0
        totalNumForInterval = 0
        for i in sbpSortedNoDupes:
            if i in interval:
                numOfHasCHDForI = len(sbp.where((sbp == i) & hasCHD).dropna())
                numOfPatientsForSBPLevelI = len(sbp.where(sbp == i).dropna())
                numHavingCHDForInterval += numOfHasCHDForI
                totalNumForInterval += numOfPatientsForSBPLevelI
        chdPercentages.append((numHavingCHDForInterval/totalNumForInterval) * 100) 
    labels = []
    for bin in bins:
        str = "{}".format(bin)
        labels.append(str)
    plt.bar(np.arange(0, len(bins)), chdPercentages, tick_label=labels)
    plt.xticks(rotation= 35, horizontalalignment='right')
    plt.xlabel(xlabel)
    axes = plt.gca()
    axes.set_ylim([0, 100])
    plt.ylabel("Percentages")
    plt.title(title)
    plt.show()
```

```python
# Part 2
# Examining the relationship between systemic blood pressure levels and coronary heart disease
title = "Percentage of patients in Systolic Blood Pressure ranges diagnosed with CHD"
generateProportionHistogram(title, "sbp", "Systolic Blood Pressure in mmHg")
```

```python
# Part 3
# Examining the relationship between serum cholesterol levels and coronary heart disease
title = "Percentage of patients in Serum Cholesterol Ranges Diagnosed with CHD"
generateProportionHistogram(title, "scl", "Serum Cholesterol in mg/dL")
```

```python
# Part 4
# Examining the relationship between age and coronary heart disease
generateProportionHistogram("Percentage of patients diagnosed with coronary heart disease by age range", "age", "Ages in years")
```

```python
# Function for getting percentages and differentiating by male or female
def generateAgeCHDProportionHistogramBySex(title: str, indVar: str, xlabel: str, sexInt: int):
    isChosenGender = male if (sexInt == 1) else female
    sbp = FHS_data[indVar]
    sbpSortedNoDupes = sbp.sort_values().drop_duplicates().dropna()
    chdPercentages = []
    binnedSortedNoDupes = pd.cut(sbpSortedNoDupes, bins=6, precision = 2)
    bins = pd.Series(binnedSortedNoDupes).drop_duplicates()
    # Calculate proportions
    for interval in bins:
        numHavingCHDForInterval = 0
        totalNumForInterval = 0
        for i in sbpSortedNoDupes:
            if i in interval:
                numOfHasCHDForI = len(sbp.where((sbp == i) & hasCHD & isChosenGender).dropna())
                numOfPatientsForSBPLevelI = len(sbp.where((sbp == i) & isChosenGender).dropna())
                numHavingCHDForInterval += numOfHasCHDForI
                totalNumForInterval += numOfPatientsForSBPLevelI
        chdPercentages.append((numHavingCHDForInterval/totalNumForInterval) * 100) 
    labels = []
    for bin in bins:
        str = "{}".format(bin)
        labels.append(str)
    plt.bar(np.arange(0, len(bins)), chdPercentages, tick_label=labels)
    plt.xticks(rotation= 35, horizontalalignment='right')
    plt.xlabel(xlabel)
    axes = plt.gca()
    axes.set_ylim([0, 100])
    plt.ylabel("Percentages")
    plt.title(title)
    plt.show()
```

```python
# Males and coronary heart disease by age
generateAgeCHDProportionHistogramBySex("Percentage of male patients diagnosed with coronary heart disease by age range", "age", "Ages in years", 1)
```

```python
# Females and coronary heart disease by age
generateAgeCHDProportionHistogramBySex("Percentage of female patients diagnosed with coronary heart disease by age range", "age", "Ages in years", 2)
```
